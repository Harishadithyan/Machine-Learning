```python
#Independent and dependent features
X=df.drop('FWI',axis=1)
Y=df['FWI']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
##Feature Slection based on correleation

X_train.corr()
```

```python
##Check for multicollinearity

plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
##threshold
correlation(X_train,0.90)
corr_features=correlation(X_train,0.85)

corr_features1=correlation(X_test,0.85)
```

```python
#dropping the features
X_train.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
X_train.columns
```

```python
X_test.drop(corr_features, axis=1, errors='ignore')
X_test.shape
X_test = X_test.drop(['BUI', 'DC'], axis=1)

X_test.columns
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
X_train_scale=Scaler.fit_transform(X_train)
X_test_scale=Scaler.transform(X_test)
```

```python
X_train_scale
```

```python
## Box plots to understand Effect of standard Scaer

plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scale)
plt.title('X_train After scaling')
```

#Model Trainning

```python
#Linear Regression Model

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
reg=LinearRegression()
reg.fit(X_train_scale,Y_train)
y_pred=reg.predict(X_test)
mae=mean_absolute_error(Y_test,y_pred)
score=r2_score(Y_test,y_pred)
print("MAE",mae)
print("r2",score)
plt.scatter(Y_test,y_pred)
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
reg=Lasso()
reg.fit(X_train_scale,Y_train)
y_pred=reg.predict(X_test)
mae=mean_absolute_error(Y_test,y_pred)
score=r2_score(Y_test,y_pred)
print("MAE",mae)
print("r2",score)
plt.scatter(Y_test,y_pred)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
reg=Ridge()
reg.fit(X_train_scale,Y_train)
y_pred=reg.predict(X_test)
mae=mean_absolute_error(Y_test,y_pred)
score=r2_score(Y_test,y_pred)
print("MAE",mae)
print("r2",score)
plt.scatter(Y_test,y_pred)
```
