```python
df_n=df.select_dtypes(int)
```

```python
df_n.head()
```

```python

X = df_n.drop(['Student_ID', 'stress_level'], axis=1) 
Y=df_n['stress_level']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,Y_train)
```

```python
y_pred=logistic.predict(X_test)
y_poba=logistic.predict_proba(X_test)
```

```python
y_pred
```

```python
y_poba
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(Y_test,y_pred))
print(accuracy_score(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```

```python
#Grid cv
penalty=['l1','l2','elasticnet']
c_values=[100,10,0,0.1,0,0.1]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']#used to find the best coefficient
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=logistic,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
score=accuracy_score(y_pred,Y_test)
print(score)
print(classification_report(y_pred,Y_test))
print(confusion_matrix(y_pred,Y_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV
model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
randomcv.fit(X_train,Y_train)
```

```python
y_pred=randomcv.predict(X_test)
```

```python
score=accuracy_score(y_pred,Y_test)
print(score)
print(classification_report(y_pred,Y_test))
print(confusion_matrix(y_pred,Y_test))
```
