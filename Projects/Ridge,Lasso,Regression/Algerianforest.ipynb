```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('dta1.csv',header=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0 # tells that it is the Bejaia Region
df.loc[122:,"Region"]=1 # Tells it is sidi-Bel Region
```

```python
df.info()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
df[['month', 'year', 'Temperature', ' RH', ' Ws']]=df[['month', 'year', 'Temperature', ' RH', ' Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.strip()
df['Classes'] = np.where(
    df['Classes'].str.contains('not fire', case=False),
    0,
    1
)
```

```python
df.head()
```

```python
df.to_csv('data_cleaned.csv',index=False)
```

```python
df1=df.drop(['day','month','year'],axis=1)
```

```python
df1.head()
```

```python
df['Classes'].value_counts()
```

```python
plt.style.use('seaborn-v0_8')
df1.hist(bins=50,figsize=(10,25))
plt.show()
```

```python
percentage=df1['Classes'].value_counts(normalize=True)*100
print(percentage)
```

```python
classlabels=['Not Fire','Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie chart of Classes')
plt.show()
```
