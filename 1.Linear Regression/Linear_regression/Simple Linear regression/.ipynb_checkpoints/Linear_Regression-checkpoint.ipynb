```python
# Independent and dependent feaures
X=df[['Weight']] #independent must be 2d array 
Y=df['Height'] #dependent feature  this can be in series or 1D array
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
#75% data  Training 25% data  Testing
#random_state is a seed value that controls how data is randomly split.It ensures to maske the same rows to go for the trrainning and testing.
```

```python
#standarization
#mean=0 and standard deviation =1
#Z_score=xi-u/sd  => Xscaled=X/
#Standardization is used to bring all features to the same scale so that no feature dominates the model and learning becomes faster and more accurate.
#Standardization converts values into how far they are from average, not their real value.
from sklearn.preprocessing  import StandardScaler
```

```python
#fit only on training data
#transform test data
scalar=StandardScaler()
scalar.fit_transform(X_train)
```

You have two features:
to make the model understand we check the data how far they are from their average
| Feature | Values            |
| ------- | ----------------- |
| Age     | 20  60           |
| Salary  | 20,000  2,00,000 |

Now the model must decide which one is important.

### What the model sees:

* Salary numbers are **very big**
* Age numbers are **very small**

So the model thinks:

> Salary is more important than age

 **Wrong conclusion**

Not because salary matters more 
but because **numbers are larger**.

---

##  Standardization fixes this

After standardization:

| Feature | Scale   |
| ------- | ------- |
| Age     | -1 to 1 |
| Salary  | -1 to 1 |

Now both are equal.

The model can **learn fairly**.

---

##  Why it is IMPORTANT

### 1 Prevents wrong learning

Without scaling:

* big numbers dominate
* model learns incorrect pattern

---

### 2 Makes training faster

Algorithms like gradient descent:

* converge faster
* fewer iterations
* less time

---

### 3 Required for distance-based models

Models like:

* KNN
* K-means
* SVM

use **distance**.

If one feature is large  distance becomes wrong.

---

### 4 Improves accuracy

Same model:

* without standardization  low accuracy
* with standardization  higher accuracy

---

##  Very simple line

> Standardization is important so that all features are treated equally by the model.

---

##  Exam / Interview one-liner

> Standardization brings all features to the same scale to prevent dominance of large-valued features and improve model performance.

---

##  One last simple example

If you measure:

* distance in **meters**
* weight in **kilograms**
* money in **rupees**

You cannot compare directly.

Standardization makes them speak **the same language**.
