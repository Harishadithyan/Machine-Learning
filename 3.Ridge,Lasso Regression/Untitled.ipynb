#Algerien Forest Fires Dataset

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the SIdi Bel-abbes region located in the nortwest of Algeria.
122 instances for each region
The period from 2012  to september 2012. The dataset includes 11 attribues and 1 output attribute.The 244 instances have been classided into fire and not fire classes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('dta1.csv',header=1)
#reomving the header or the row  one using header=1
```

The dataset is converted into two sets based on Region from 122 index , we can make a new column
based on the Region
1: "Bejaia Region Dataset"
2: "Sidi-Bel Abbes Region Dataset"
Add new column with region

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df[df.isnull().any(axis=1)]
```

```python
#Removing the null values 
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().any(axis=1)
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in the columns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
## Change the required columns as the req data type
df[['month','day','year','Temperature','RH','Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
## changeing the other columns to float datatype

objects=[feature for feature in df.columns if df[feature].dtype == 'O']
print(objects)
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
## Lets save the cleaned dataset
df.to_csv('data_cleaned.csv',index=False)
```

```python
#Exploratory Data Analysis
#dropping dat,mont,year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
## Encoding the categories in classes only once exe cuted

df_copy['Classes'] = df_copy['Classes'].astype(str)

df_copy['Classes'] = np.where(
    df_copy['Classes'].str.contains('1'),
    0,
    1
)
df_copy['Classes'] = df_copy['Classes'].astype(int)
```

```python
#converting some data to fire
df_copy.iloc[150:240, df_copy.columns.get_loc('Classes')] = 1
```
